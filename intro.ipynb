The fields of an object are the data items we want to associate together. For example, if I want to track book titles/authors, I can use a list of tuples:

```python
books = [
    {'title':'Gridlinked', 'author':'Neal Asher'},
    {'title':'Startide Rising', 'author':'David Brin'}
]
```

```python
for b in books:
    print(f"{b['author']}: {b['title']}")
```

```python
s = 0
def foo():
    s = 99
    print(s)
foo()
print(s)
```

```python
s = 0
def foo():
    global s
    s = 99
    print(s)
foo()
print(s)
```

```python
s = 1
def foo():
    print(s + 100)
foo()
print(s)
```

```python
s = 1
def foo():
    s = s + 100
    print(s)
foo()
print(s)
```
